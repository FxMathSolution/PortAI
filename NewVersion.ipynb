```python
import pandas as pd
df1=pd.read_csv('EURUSD-M1.csv',header=None,index_col=False)
df60=pd.read_csv('EURUSD-H1.csv',header=None,index_col=False)
df60.columns=['date','time','open','high','low','close','volume']
df1.columns=['date','time','open','high','low','close','volume']
df60.head()
```

```python
df60['date']=df60['date']+" "+df60['time']
df60=df60.drop(columns=['time'])
df1['date']=df1['date']+" "+df1['time']
df1=df1.drop(columns=['time'])
```

```python
window=12
digit=10000
sl=70
tp=100
```

```python
def findIndex(dDate):
  return df1.index[df1['date'] == dDate].tolist()

def calSLTP(hdf):
  hi=hdf['high'].max()
  lo=hdf['low'].min()
  ihi=hdf.index[hdf['high'] == hdf['high'].max()].tolist()[0]
  ilo=hdf.index[hdf['low'] == hdf['low'].min()].tolist()[0]
  hi=(hi-hdf['open'].iloc[0])*digit
  lo=(lo-hdf['open'].iloc[0])*digit
  if abs(hi)>abs(lo) and ihi<ilo and abs(lo)<sl: return([lo/tp,hi/tp])
  elif abs(hi)<abs(lo) and ihi>ilo and abs(hi)<sl: return([hi/tp,lo/tp])
  elif abs(hi)>abs(lo) and ihi>ilo: return([hi/tp,lo/tp])
  elif abs(hi)<abs(lo) and ihi<ilo: return([lo/tp,hi/tp])
  else: return [0,0]


```

```python
alow=[]
ahigh=[]
for x in range(100,220):
  adf=df60[x:x+window]
  sInd=findIndex(adf['date'].iloc[0])
  if sInd!=[]: sInd=sInd[0]
  else: continue
  eInd=sInd+(window*60) 
  q=calSLTP(df1[sInd:eInd])
  print(q)

#print(min(alow),max(ahigh))
```
